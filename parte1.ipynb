```python
import requests

from bs4 import BeautifulSoup
```

### Enviar petición HTTP para extraer código fuente de la Página Web

```python
page = requests.get('https://www.gonvill.com.mx/libros-de/libros-para-todos-10/')
#page = requests.get('https://rinacional.tecnm.mx/handle/TecNM/4484')
#page = requests.get('https://vocero.uach.mx/index.php/Estudios_Informacion/article/view/1284')
soup = BeautifulSoup(page.text, 'html.parser')
print(soup.prettify())
```

### Analizar codigo fuente

```python
titulos = []
titulo_items = soup.find_all('div', class_="portada")
for items in titulo_items:
    titulo = items.a.img
    pos = titulo['title'].find("(")
    #print(titulo['title'])
    titulos.append(titulo['title'][0:pos].strip())
print(len(titulos))
titulos
```

```python
ligas = []
ligas_items = soup.find_all('div', class_="portada")
for items in titulo_items:
    titulo = items.a
    #print('https://www.gonvill.com.mx'+titulo['href'])
    ligas.append('https://www.gonvill.com.mx'+titulo['href'])
print(len(ligas))
ligas
```

```python
images = []

images_items = soup.find_all('img', class_="foto")
#print(images_items)
for item in images_items:
    image = item['src']
    #image = item['class']
    #print(image)
    images.append(image)
print(len(images))
#print(images_items)
images
```

```python
precios = []
precio_items = soup.find_all('p',class_="precio")
#print(precio_items)
for item in precio_items:
    # accedemos a su etiqueta a
    #cprecio = item.text.split('Â')
    cprecio = item.text.strip()
    pos2 = cprecio.find(".")
    precios.append(cprecio[1:pos2+3])
len(precios)
precios
```

```python
creador = []
creador_items = soup.find_all('dl', class_="dublincore")
for items in creador_items:
    autor = items.find(class_="creator").text.strip()
    #print(autor)
    creador.append(autor)
print(len(creador))
creador
```

```python
# como guardar todos estos arreglos en un dataFrame
import pandas as pd
data = {
    'Titulo': titulos,
    'Creadores': creador,
    'Imagenes': images,
    'Ligas': ligas,
    'Precios': precios
}
df = pd.DataFrame(data)
df
```

```python
import datetime
fecha_actual = datetime.datetime.now().strftime("%d-%m-%Y")
df.to_csv(f"dataset/Gonvill-{fecha_actual}.csv",index=False)
```
